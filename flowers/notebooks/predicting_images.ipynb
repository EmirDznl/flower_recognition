```python
import numpy as np
import tensorflow as tf
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import img_to_array, load_img
import matplotlib.pyplot as plt
```

```python
flower_categories = ['daisy', 'dandelion' , 'rose', 'sunflower' , 'tulip']
```

```python
model = tf.keras.models.load_model('flower_model.h5')
```

```python
img_dir = "../sample_data/test/Image_5.jpg"
```

```python
test_image = tf.keras.utils.load_img(img_dir, target_size=(224,224))
show_image = tf.keras.utils.load_img(img_dir)

test_image = tf.keras.utils.img_to_array(test_image)

test_image = np.expand_dims(test_image, axis=0)

result = model.predict(test_image)
```

```python
result
```

```python
def type_finder(arr):
    index = np.where(arr == 1)[1]
    if index == 0:
        flower = "papatya"
    elif index == 1:
        flower = "karahidiba"
    elif index == 2:
        flower = "gül"
    elif index == 3:
        flower = "ayçiçeği"
    elif index == 4:
        flower = "lale"
    print("Bu bir {} fotoğrafıdır".format(flower))
```

```python
type_finder(result)
plt.imshow(show_image)
plt.axis("off")
plt.show()
```
